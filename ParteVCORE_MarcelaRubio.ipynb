1. **Heatmap:** Realiza un análisis de correlación y crea un mapa de calor para visualizar las relaciones entre las variables.
* obs: HeatMap de variables numéricas originales, no consideré las que se agregaban en los core anteriores

```python
# Mapa de calor de correlación entre variables numéricas.
corr = df2[["age", "quantity", "price_per_unit", "total_amount"]].corr()
sns.heatmap(corr, annot=True, cmap="coolwarm", linewidths=0.5)
plt.title("Matriz de correlación")
plt.show()

```

* **total_amount vs price_per_unit** = 0.85 = Correlación muy fuerte y positiva. A mayor precio por unidad, mayor monto total. Es lógico porque el monto total depende del precio.
* **total_amount vs quantity** = 0.37 = Correlación moderada y positiva. Más unidades, mayor monto total.
* **age con cualquier otra** = 0 = Correlación muy baja o nula. La edad no parece estar relacionada directamente con la cantidad, precio ni monto.
* **quantity vs price_per_unit** = 0.018 = Prácticamente sin relación.

**En conclusión:**
1. La variable total_amount está fuertemente influenciada por price_per_unit y moderadamente por quantity.

2. La edad (age) no tiene una relación aparente con las variables de compra, al menos no linealmente.


# **Subplots para comparar**
Crea subplots para comparar diferentes variables clave.
Añade cuadrículas, leyendas, anotaciones y flechas a tus gráficos para mejorar la claridad y la información.


```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# figura y ejes
fig, axes = plt.subplots(2, 2, figsize=(14, 10))
fig.suptitle('Comparación de variables clave', fontsize=16)

# Subplot 1: Edad vs Monto total
sns.scatterplot(ax=axes[0, 0], data=df2, x='age', y='total_amount', color='teal')
axes[0, 0].set_title('Edad vs Monto total')
axes[0, 0].grid(True)

# Anotación (por ejemplo, si alguien gastó mucho a edad joven)
punto = df2[df2['total_amount'] == df2['total_amount'].max()].iloc[0]
axes[0, 0].annotate('Máximo gasto',
                    xy=(punto['age'], punto['total_amount']),
                    xytext=(punto['age'] + 5, punto['total_amount'] - 200),
                    arrowprops=dict(facecolor='black', shrink=0.05))

# Subplot 2: Precio por unidad vs Monto total
sns.scatterplot(ax=axes[0, 1], data=df2, x='price_per_unit', y='total_amount', color='darkorange')
axes[0, 1].set_title('Precio unidad vs Monto total')
axes[0, 1].grid(True)

# Subplot 3: Cantidad vs Monto total
sns.scatterplot(ax=axes[1, 0], data=df2, x='quantity', y='total_amount', color='purple')
axes[1, 0].set_title('Cantidad vs Monto total')
axes[1, 0].grid(True)

# Subplot 4: Edad vs Cantidad
sns.scatterplot(ax=axes[1, 1], data=df2, x='age', y='quantity', hue='gender', palette='Set1')
axes[1, 1].set_title('Edad vs Cantidad (por género)')
axes[1, 1].grid(True)
axes[1, 1].legend(title='Género')

plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()
```

# **Edad vs Monto total (Gráfico superior izquierdo)**
**Tendencia:** No se observa una relación clara entre la edad y el monto total gastado.

**Observación destacada:** Se ha marcado un punto como “Máximo gasto”, alrededor de los 44 años con un monto total cercano a 2000.

**Conclusión:** Personas de distintas edades realizan gastos altos, pero no se concentra en un grupo etario específico.

# **Precio unidad vs Monto total (Gráfico superior derecho)**
**Tendencia:** A medida que el precio por unidad aumenta, también lo hace el monto total. Esto es esperado, ya que un precio mayor por unidad impacta directamente en el total.

**Distribución:** Se observan precios agrupados en ciertos valores específicos, lo que podría indicar que los productos tienen precios fijos (ej: 50, 300, 500).

**Conclusión:** Hay una relación positiva clara entre el precio por unidad y el monto total.

# **Cantidad vs Monto total (Gráfico inferior izquierdo)**
**Tendencia:** Cuanto mayor es la cantidad comprada, mayor puede ser el monto total. Sin embargo, también hay mucha variabilidad (monto bajo incluso con cantidades altas).

**Distribución:** Las cantidades están limitadas a valores bajos (entre 1 y 4), lo que puede estar relacionado con una política de compra o comportamiento del cliente.

**Conclusión:** El total gastado depende de la cantidad, pero también de otros factores como el precio por unidad.

# **Edad vs Cantidad (por género) (Gráfico inferior derecho)**
**Colores:** Azul para mujeres, rojo para hombres.

**Tendencia:** La cantidad comprada no varía significativamente con la edad.

**Distribución:** Tanto hombres como mujeres están distribuidos de manera bastante homogénea en todas las edades y cantidades.

**Conclusión:** No parece haber una diferencia significativa entre géneros o edades en cuanto a las cantidades compradas.
